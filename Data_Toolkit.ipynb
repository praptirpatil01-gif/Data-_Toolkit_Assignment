Q2.How does broadcasting work in NumPy?

Ans:- Broadcasting in NumPy allows arrays of different shapes to participate in arithmetic operations without needing explicit resizing. NumPy automatically expands the smaller array to match the shape of the larger one based on broadcasting rules. This makes operations faster, memory-efficient, and avoids writing loops manually.

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

result = a + b
print(result)
```

Q6. What are the differences between NumPy arrays and Python lists?

Ans:-
**NumPy Arrays:-**

 * NumPy arrays store elements of the same data type (homogeneous), which makes them more memory-efficient.

 * They support vectorized operations, meaning mathematical operations can be applied to the entire array without loops.

 * NumPy arrays are faster because they are stored in continuous memory and use optimized computation.

**Python Lists:-**

* Python lists can store different data types (heterogeneous) in the same structure.

 * Mathematical operations are not applied directly; you must use loops or list comprehension.

 * They are slower and use more memory because each element is stored separately.

Q17.Explain the difference between apply() and map() in Pandas.

Ans:-
**map() Function:**

* Used mainly with Series (single column).

* Applies a function element-wise to each value.

* Best for simple transformations or replacements.

```python
import pandas as pd

df = pd.DataFrame({"Marks": [50, 60, 70]})

result = df["Marks"].map(lambda x: x + 5)
print(result)
```

**apply() Function:-**

* Works on Series and DataFrames.

* Can apply a function to rows or columns.

* Used when the operation is more complex.

```python
import pandas as pd

df = pd.DataFrame({"Marks": [50, 60, 70]})

result = df["Marks"].apply(lambda x: x + 5)
print(result)
```

Q4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({"A":[1,2,3], "B":[2,4,6], "C":[3,6,9]})
corr = df.corr()

sns.heatmap(corr, annot=True)
plt.show()
```
